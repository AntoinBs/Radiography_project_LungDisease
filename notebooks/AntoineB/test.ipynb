```python
import pandas as pd
import numpy as np

import cv2
from pathlib import Path

from sklearn.metrics import classification_report

from Libraries.get_absolute_path import get_absolute_path

from tensorflow.keras.models import load_model
```

```python
model = load_model(get_absolute_path(r"models\model_xception.keras"))
```

```python
class_names = ['COVID', 'LUNG OPACITY', 'NORMAL', 'VIRAL PNEUMONIA']
```

```python
#from keras.applications.inception_v3 import preprocess_input

def prepare_image(image_path, target_size=(299, 299)):
    image = cv2.imread(image_path)
    #image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convertir en RGB
    image = cv2.resize(image, target_size)  # Redimensionner
    #image = image / 127.5 - 1 # Normaliser
    return np.expand_dims(image, axis=0)  # Ajouter batch dimension
```

```python
def predict_images(model, path):
    predictions = []
    for label in path.glob("*"):
        path = Path(path, label)
        for img_file in path.glob("*.png"):
            print(img_file.name)
            image = prepare_image(img_file)
            prediction = np.argmax(model.predict(image), axis=1)
            prediction = class_names[prediction[0]]
            predictions.append([img_file, label.name, prediction])
    return pd.DataFrame(predictions, columns=['FILE', 'TRUE_LABEL', 'PREDICTED_LABEL'])
```

```python
path = get_absolute_path(r"data\processed\train_test_split\test")
predictions = predict_images(model, path)
```
