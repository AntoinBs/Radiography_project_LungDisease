```python
import pandas as pd
import numpy as np
import cv2
```

```python
# Echantillon d'images
url_images = {
    'normal_1' : r'..\data\raw\Normal\images\Normal-10.png',
    'normal_2' : r'..\data\raw\Normal\images\Normal-989.png',
    'normal_3' : r'..\data\raw\Normal\images\Normal-4513.png',
    "covid_1" : r'..\data\raw\COVID\images\COVID-233.png',
    "covid_2" : r'..\data\raw\COVID\images\COVID-909.png',
    "covid_3" : r'..\data\raw\COVID\images\COVID-1761.png',
    "viral_pneumonia_1" : r'..\data\raw\Viral Pneumonia\images\Viral Pneumonia-333.png',
    "viral_pneumonia_2" : r'..\data\raw\Viral Pneumonia\images\Viral Pneumonia-1070.png',
    "viral_pneumonia_3" : r'..\data\raw\Viral Pneumonia\images\Viral Pneumonia-1338.png',
    "lung_opacity_1" : r'..\data\raw\Lung_Opacity\images\Lung_Opacity-15.png',
    "lung_opacity_2" : r'..\data\raw\Lung_Opacity\images\Lung_Opacity-641.png',
    "lung_opacity_3" : r'..\data\raw\Lung_Opacity\images\Lung_Opacity-1839.png',
}
```

```python
images = {} # dic qui contiendra les images
for key in url_images.keys():
    name = "img_" + key # Le nom de l'image sera img_ + le nom donn√© dans url_images
    images[name] = cv2.imread(url_images[key], cv2.IMREAD_GRAYSCALE) # lecture des images en noir et blanc
```

```python
import matplotlib.pyplot as plt

# Affichage de l'√©chnantillon d'images
plt.figure(figsize=(15,15))
for i, img_name in zip(range(len(images)), images.keys()):
    plt.subplot(4, 3, i+1)
    plt.imshow(images[img_name], cmap='gray')
    plt.title(img_name)
    plt.xticks([])
    plt.yticks([])
```
