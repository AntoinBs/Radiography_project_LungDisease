Dans cette celulle nous sélectionnons une image aléatoirement, la traitons et affichons les décisions de notre modèle dans des listes.

```python
import random
import os
from pathlib import Path
import cv2
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications.inception_v3 import preprocess_input

# Initialisation des listes pour stocker les prédictions et les vérités terrain
predictions = []
realities = []

for i in range(100):
    # Sélection aléatoire d'une image non masquée
    parent_folder = '../data/raw'
    labels = ['COVID', 'Lung_Opacity', 'Normal', 'Viral_Pneumonia']
    selected_label = random.choice(labels)
    file_selected = random.choice(os.listdir(Path(parent_folder, selected_label, 'images')))
    file_path = Path(parent_folder, selected_label, 'images', file_selected)

    # # Afficher l'image sélectionnée et son label
    img = cv2.imread(str(file_path), cv2.IMREAD_GRAYSCALE)
    # plt.figure(figsize=(10, 5))
    # plt.imshow(img, cmap='gray')
    # plt.title(f"Image sélectionnée : {file_selected}")
    # plt.show()  

    # Convertion vers un tensor RGB
    img = np.resize(img, (299, 299, 1))
    img = tf.convert_to_tensor(img, dtype=tf.float16)
    img_rgb = tf.image.grayscale_to_rgb(img, name=None)

    # Preprocessing de l'image
    img_rgb = preprocess_input(img_rgb)
    # Reshape pour avoir une forme correcte
    img_rgb = tf.expand_dims(img_rgb, axis=0)  # Forme (1, 299, 299, 3)


    print("Image Tensor Shape:", img_rgb.shape)
    print("Sample Pixel Range:", img_rgb.numpy().min(), img_rgb.numpy().max())

    # Prediction du modèle et comparaison avec la réalité
    y_predict = model.predict(img_rgb)
    predicted_class = labels[int(y_predict.argmax(axis=1))]
    predictions.append(predicted_class)
    realities.append(selected_label)

# Affichage des résultats
print("Prédictions :", predictions)
print("Vérité :", realities)

df=pd.DataFrame()
df['Predictions']=predictions
df['Reality']=realities
```
